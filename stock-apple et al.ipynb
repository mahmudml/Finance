```python
import pandas 
import datetime
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
!pip install yfinance
import yfinance as yf
start = '2014-02-09'
end ='2022-02-09'
aapl = yf.download('AAPL',start,end)
msft = yf.download('MSFT',start,end)
tsla = yf.download('TSLA',start,end)
aapl['Open'].plot(label = 'Apple', figsize = (15,7))
msft['Open'].plot(label = 'Micsosoft')
tsla['Open'].plot(label = 'Tesla')
pyplot.title('Opening Share Prices of Stock Traded')
pyplot.legend()
```

```python
aapl['Volume'].plot(label='Apple', figsize=(15,7))
msft['Volume'].plot(label='Micrsosoft')
tsla['Volume'].plot(label='Tesla')
pyplot.title('Volume of Stock Traded')
pyplot.legend()
```

```python
# calculation of martket cap
aapl['MarketCap']=aapl['Volume']*aapl['Close']
msft['MarketCap']=msft['Volume']*msft['Close']
tsla['MarketCap']=tsla['Volume']*tsla['Close']
aapl['MarketCap'].plot(label='Apple', figsize=(15,7))
msft['MarketCap'].plot(label='Microsoft')
tsla['MarketCap'].plot(label='Tesla')
pyplot.title('Market Cap of Traded Stocks')
pyplot.legend()
```
