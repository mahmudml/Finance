```python
import pandas 
import datetime
from matplotlib import pyplot
from pandas.plotting import scatter_matrix
!pip install yfinance
import yfinance as yf
start = '2014-02-09'
end ='2022-02-08'
aapl = yf.download('AAPL',start,end)
msft = yf.download('MSFT',start,end)
tsla = yf.download('TSLA',start,end)
aapl['Close'].plot(label = 'Apple', figsize = (15,7))
msft['Close'].plot(label = 'Micsosoft')
tsla['Close'].plot(label = 'Tesla')
pyplot.title('Closing Share Prices of Stock Traded')
pyplot.legend()
```

```python
# calculation of Volume
aapl['Volume'].plot(label='Apple', figsize=(15,7))
msft['Volume'].plot(label='Micrsosoft')
tsla['Volume'].plot(label='Tesla')
pyplot.title('Volume of Stock Traded')
pyplot.legend()

```

```python
# calculation of market cap
aapl['MarketCap']=aapl['Volume']*aapl['Close']
msft['MarketCap']=msft['Volume']*msft['Close']
tsla['MarketCap']=tsla['Volume']*tsla['Close']
aapl['MarketCap'].plot(label='Apple', figsize=(15,7))
msft['MarketCap'].plot(label='Microsoft')
tsla['MarketCap'].plot(label='Tesla')
pyplot.title('Market Cap of Traded Stocks')
pyplot.legend()
```

```python
#calculation of apple stock moving averages
aapl['MA50']=aapl['Close'].rolling(50).mean()
aapl['MA200']=aapl['Close'].rolling(200).mean()
aapl['Close'].plot(label='Closing Price', figsize=(15,7))
aapl['MA50'].plot(label='50 days moving average')
aapl['MA200'].plot(label='200 days moving average')
pyplot.title('Apple stock closing prices comparison with moving averages')
pyplot.legend()
```

```python
msft['MA50']=msft['Close'].rolling(50).mean()
msft['MA200']=msft['Close'].rolling(200).mean()
msft['Close'].plot(label='Closing Price', figsize=(15,7))
msft['MA50'].plot(label='50 days moving average')
msft['MA200'].plot(label='200 days moving average')
pyplot.title('Microsoft stock closing prices comparison with moving averages')
pyplot.legend()
```

```python
tsla['MA50']=tsla['Close'].rolling(50).mean()
tsla['MA200']=tsla['Close'].rolling(200).mean()
tsla['Close'].plot(label='Closing Price', figsize=(15,7))
tsla['MA50'].plot(label='50 days moving average')
tsla['MA200'].plot(label='200 days moving average')
pyplot.title('Tesla stock closing prices comparison with moving averages')
pyplot.legend()
```
